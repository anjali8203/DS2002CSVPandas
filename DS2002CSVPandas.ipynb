```python
import pandas as pd
# 1. reading the CSV file

raw_data = pd.read_csv('/content/acc_players-2324F.csv', sep=';', encoding='latin1', header=None)
basketball_data = raw_data.drop(0)

basketball_data = pd.read_csv('/content/acc_players-2324F.csv', sep=',', encoding='latin1', skiprows=1)

display(basketball_data.head())
```

```python
basketball_data.info()
```

```python
basketball_data.describe()
```

```python
# prompt: calculate the total points scored by all players combined

total_points = basketball_data['PTS'].sum()
print(f"Total points scored by all players: {total_points}")
```

```python
# calculate total points scored by all students combined
max_val = basketball_data.loc[basketball_data['MP'].idxmax()]
print(f"Max minutes: {max_val['Player']}")
```

```python
# top 5 players with most rebounds
top_5_rebounds = basketball_data.sort_values(by='TRB', ascending=False).head(5)

print("Top 5 players with the most rebounds:")
print(top_5_rebounds[['Player', 'TRB']])
```

```python
# new dataframe with players who played more than 500 times
players_played_more_than_500 = basketball_data[basketball_data['MP'] > 500]
players_played_more_than_500
```

```python
# determine player with highest total assists
player_with_highest_assists = players_played_more_than_500.loc[players_played_more_than_500['AST'].idxmax()]
player_with_highest_assists
```

```python
# determine the top three players with the most assists in the league
top_three_assists = basketball_data.nlargest(3, 'AST')
top_three_assists
```

```python
# determine the top three shot blockers in the league from the new dataset
top_three_shot_blockers = basketball_data.nlargest(3, 'BLK')
top_three_shot_blockers
```

```python
# SCHOOL BASED ANALYSIS

# group players by school and calculate total points scored per school
school_points = basketball_data.groupby('School')['PTS'].sum()
school_points.sort_values(ascending=False)
```

```python
# group players by school and calculate total assists (AST) per school
school_assists = basketball_data.groupby('School')['AST'].sum()
school_assists.sort_values(ascending=False)
```

```python
# Sort the schools by total points scored and display the top 3 schools.
top_3_schools_points = school_points.nlargest(3)
top_3_schools_points
```

```python
import matplotlib.pyplot as plt
top_5_players_points = basketball_data.nlargest(5, 'PTS')

plt.bar(top_5_players_points['Player'], top_5_players_points['PTS'], color='skyblue')
plt.xlabel('Player')
plt.ylabel('Total Points Scored (PTS)')
plt.title('Top 5 Players by Total Points Scored')
plt.xticks(rotation=45, ha='right')

plt.show()
```

```python
# BONUS

correlation_coefficient = basketball_data['FG%'].corr(basketball_data['WS'])


plt.scatter(basketball_data['FG%'], basketball_data['WS'], color='green')
plt.xlabel('Field Goal Percentage (FG%)')
plt.ylabel('Win Shares (WS)')
plt.title(f'Correlation between FG% and WS: {correlation_coefficient:.2f}')
plt.grid(True)

plt.show()

print(f'Correlation coefficient: {correlation_coefficient:.2f}')
```
